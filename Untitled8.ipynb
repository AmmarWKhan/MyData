```python
# Step 1: Connect to the ZKTeco device and retrieve attendance data
import sys
import os
from zk import ZK
```

```python
# Connect to ZKTeco device
zk_ip_address = '182.176.178.62'

# Connect to ZKTeco device
zk = ZK(zk_ip_address, port=4370, timeout=5, password=0, force_udp=False, ommit_ping=False)
conn = zk.connect()
```

```python
# Get attendances
attendances = conn.get_attendance()
```

```python
import pandas as pd
import re

# For example, attendances = conn.get_attendance()

# Define regular expressions to extract information
attendance_pattern = r"<Attendance>: (\d+) : (\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}) \((\d+), (\d+)\)"

# Extract data and create a list of dictionaries
data = []

for attendance_string in attendances:
    # Convert the attendance object to a string
    attendance_string = str(attendance_string)
    
    attendance_match = re.search(attendance_pattern, attendance_string)

    if attendance_match:
        user_id = int(attendance_match.group(1))
        datetime_str = attendance_match.group(2)
        verify_state_id = int(attendance_match.group(3))
        punch_type_id = int(attendance_match.group(4))
        
        # Split the datetime string into date and time
        date, time = datetime_str.split(' ')

        data.append({
            "user_id": user_id,
            "punch_date":date,
        "punch_time":date + ' ' +time,
            "verify_state_id": verify_state_id,
            "punch_type_id": punch_type_id
        })

# Create a DataFrame from the list of dictionaries
df_attendance = pd.DataFrame(data)

# Display the DataFrame
print(df_attendance) 
```

```python
import pymysql
from datetime import datetime

# Connect to MySQL
db = pymysql.connect(host='192.168.100.38', user='rd_user', password='Rd1234', database='public_ip_db_test')
cursor = db.cursor()

# Step 1: Extract data for the specific office_id
office_id = office_id  # Replace with the extracted office_id
data_to_insert = df_attendance_detail[df_attendance_detail['office_id'] == office_id]

# Step 2: Prepare data for insertion
for index, row in data_to_insert.iterrows():
    # Extract values from the row
    office_id = row['office_id']
    uid = row['emp_dev_uid']
    emp_id = row['emp_id']
    first_name = row['first_name']
    last_name = row['last_name']
    punch_date = row['punch_date']
    punch_time = row['punch_time']
    punch_type_id = row['punch_type_id']
    punch_type_display = row['punch_type_display']
    verify_state_id = row['verify_state_id']
    verify_state_display = row['verify_state_display']

    # Generate current datetime for upload_time, created_at, updated_at
    current_datetime = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

    # Define the SQL query to check for duplicates
    check_duplicate_sql = f"SELECT COUNT(*) FROM attendances WHERE emp_id = {emp_id} AND punch_date = '{punch_date}' AND punch_time = '{punch_time}'"

    try:
        # Execute the SQL query to check for duplicates
        cursor.execute(check_duplicate_sql)
        if cursor.fetchone()[0] == 0:  # If no duplicate is found
            # Define the SQL query for insertion
            insert_sql = f"""
                INSERT INTO attendances (office_id, uid, emp_id, first_name, last_name, punch_date, punch_time, punch_type_id, punch_type_display, verify_state_id, verify_state_display, upload_time, created_at, updated_at, email)
                VALUES ({office_id}, '{uid}', {emp_id}, '{first_name}', '{last_name}', '{punch_date}', '{punch_time}', {punch_type_id}, '{punch_type_display}', {verify_state_id}, '{verify_state_display}', '{current_datetime}', '{current_datetime}', '{current_datetime}', NULL)
            """
            
            # Execute the SQL query for insertion
            cursor.execute(insert_sql)
            # Commit changes to the database
            db.commit()
            print(f"Inserted data for emp_id {emp_id}")
        else:
            print(f"Data for emp_id {emp_id} already exists, skipping insertion.")
    except Exception as e:
        # Rollback if an error occurs
        db.rollback()
        print(f"Error inserting data for emp_id {emp_id}: {e}")

# Step 3: Close the connection
db.close()
```
