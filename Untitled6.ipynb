```python
import pandas as pd
import re

# Define regular expressions to extract information
uid_pattern = r"uid:(\d+)"
name_pattern = r"name:(.*?)(?: user_id:\d+|$)"  
user_id_pattern = r"user_id:(\d+)"

# Extract data and create a list of dictionaries
data = []

# Assuming df_users is your dataframe
for index, row in df_users.iterrows():
    user_string = str(row[0])  # Assuming the user data is in the first column
    
    uid_match = re.search(uid_pattern, user_string)
    name_match = re.search(name_pattern, user_string)
    user_id_match = re.search(user_id_pattern, user_string)

    if uid_match and name_match and user_id_match:
        uid = int(uid_match.group(1))
        full_name = name_match.group(1).strip()  # Remove leading/trailing spaces
        user_id = int(user_id_match.group(1))

        data.append({'uid': uid, 'name': full_name, 'user_id': user_id})

# Create a DataFrame from the list of dictionaries
df_users = pd.DataFrame(data)

# Display the DataFrame
print(df_users)
```

```python
import pandas as pd
import re

# For example, attendances = conn.get_attendance()

# Define regular expressions to extract information
attendance_pattern = r"<Attendance>: (\d+) : (\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}) \((\d+), (\d+)\)"

# Extract data and create a list of dictionaries
data = []

for attendance_string in attendances:
    # Convert the attendance object to a string
    attendance_string = str(attendance_string)
    
    attendance_match = re.search(attendance_pattern, attendance_string)

    if attendance_match:
        user_id = int(attendance_match.group(1))
        datetime_str = attendance_match.group(2)
        verify_state_id = int(attendance_match.group(3))
        punch_type_id = int(attendance_match.group(4))
        
        # Split the datetime string into date and time
        date, time = datetime_str.split(' ')

        data.append({
            "user_id": user_id,
            "punch_date":date,
        "punch_time":date + ' ' +time,
            "verify_state_id": verify_state_id,
            "punch_type_id": punch_type_id
        })

# Create a DataFrame from the list of dictionaries
df_attendance = pd.DataFrame(data)

# Display the DataFrame
print(df_attendance) 
```
